# Advanced indexing

```python
import pandas as pd
import numpy as np
```

### Changing index of a DataFrame

```python
sales = pd.read_csv('sales/sales.csv', index_col='month')
sales.head()
```

```python
new_idx = [i.upper() for i in sales.index]
sales.index = new_idx
sales
```

### Changing index name labels

```python
sales.index.name = 'MONTHS'
sales.columns.name = 'PRODUCTS'
sales
```

### Building an index, then a DataFrame

```python
sales = pd.read_csv('sales/sales.csv')
sales.head()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales.index = months
sales
```

### Extracting data with a MultiIndex, seting index

```python
sales = pd.read_csv('sales/sales.csv')
sales.head()
```

```python
sales['month'] = [1,2,1,2,1,2]
sales['state'] = ['CA','CA','NY','NY','TX','TX']
sales = sales.set_index(['state','month'])
sales = sales.sort_index()
sales
```

```python
sales.loc[['CA','TX']]
```

```python
sales['CA':'TX']
```

```python
sales.loc['CA']
```

### Indexing multiple levels of a MultiIndex

```python
sales.loc['NY',1]
```

```python
sales.loc[(['CA','TX'],2),:]
```

```python
sales.loc[(slice(None),2),:]
```
