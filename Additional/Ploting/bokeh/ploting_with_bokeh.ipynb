# Plotting with Bokeh

```python
from bokeh.io import show, output_notebook
#from bokeh.io import output_file
from bokeh.plotting import figure
import numpy as np
import pandas as pd
output_notebook()
```

```python
N = 4000
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radii = np.random.random(size=N) * 1.5
colors = [
    "#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip(50+2*x, 30+2*y)
]

TOOLS="hover,crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,"

p = figure(tools=TOOLS)

p.scatter(x, y, radius=radii,
          fill_color=colors, fill_alpha=0.6,
          line_color=None)

# output_file('bokeh.html')
show(p)
```

### A simple scatter plot

```python
fertility = pd.read_csv('literacy_birth_rate.csv')
```

```python
fertility.dropna(inplace=True)
fertility['fertility'] = pd.to_numeric(fertility['fertility'])
fertility['female literacy'] = pd.to_numeric(fertility['female literacy'])
fertility.info()
```

```python
fertility_list = fertility['fertility'].values.tolist()
female_literacy = fertility['female literacy'].values.tolist()
```

```python
p = figure(x_axis_label='fertility (children per woman)', y_axis_label='female_literacy (% population)')
p.circle(fertility_list, female_literacy)

show(p)
```

### A scatter plot with different shapes

```python
fertility_africa = fertility[fertility['Continent'] == 'AF']['fertility'].values.tolist()
female_literacy_africa = fertility[fertility['Continent'] == 'AF']['female literacy'].values.tolist()
fertility_latinamerica = fertility[fertility['Continent'] == 'LAT']['fertility'].values.tolist()
female_literacy_latinamerica = fertility[fertility['Continent'] == 'LAT']['female literacy'].values.tolist()
```

```python
p = figure(x_axis_label='fertility', y_axis_label='female_literacy (% population)')
```

```python
p.circle(fertility_latinamerica,female_literacy_latinamerica)
p.x(fertility_africa, female_literacy_africa)

show(p)
```

### Customizing your scatter plots

```python
p = figure(x_axis_label='fertility (children per woman)', y_axis_label='female_literacy (% population)')

p.circle(fertility_latinamerica, female_literacy_latinamerica, color='blue', size=10, alpha=0.8)
p.circle(fertility_africa, female_literacy_africa, color='red', size=10, alpha=0.8)

show(p)
```

### Hover glyphs

```python
from bokeh.models import HoverTool
```

```python
df = pd.read_csv('glucose.csv')
df.head()
```

```python
p = figure()
```

```python
p.circle(df.index.values, df.glucose.values, size=10,
         fill_color='grey', alpha=0.1, line_color=None,
         hover_fill_color='firebrick', hover_alpha=0.5,
         hover_line_color='white')

hover = HoverTool(tooltips=None, mode='vline')

p.add_tools(hover)

show(p)
```

### Colormapping

```python
df = pd.read_csv('auto-mpg.csv')
```

```python
from bokeh.models import CategoricalColorMapper
```

```python
source = ColumnDataSource(df)
p = figure()
```

```python
color_mapper = CategoricalColorMapper(factors=['Europe', 'Asia', 'US'],
                                      palette=['red', 'green', 'blue'])
```

```python
p.circle('weight', 'mpg', source=source,
            color=dict(field='origin', transform=color_mapper),
            legend='origin')
show(p)
```
