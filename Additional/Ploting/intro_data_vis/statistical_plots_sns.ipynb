# Statistical plots with Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('auto-mpg.csv')
df.head()
```

### Simple linear regressions

```python
sns.lmplot(x='weight',y='hp', data=df)
plt.show()
```

### Plotting residuals of a regression

```python
sns.residplot(x='hp',y='mpg', data=df, color='green')
plt.show()
```

### Higher-order regressions

```python
plt.scatter(df['weight'], df['mpg'], label='data', color='red', marker='o')
sns.regplot(x='weight',y='mpg',data=df,color='blue', scatter=None, label='order 1')
sns.regplot(x='weight',y='mpg',data=df,color='green', scatter=None, label='order 2', order=2)

plt.legend(loc='upper right')
plt.show()
```

### Grouping linear regressions by hue

```python
sns.lmplot(x='weight',y='hp',data=df,hue='origin',palette="Set1")
plt.savefig('reg.png')
plt.show()
```

### Grouping linear regressions by row or column

```python
sns.lmplot(x='weight',y='hp',data=df,row='origin')
plt.show()
```

### Constructing strip plots

```python
plt.subplot(2,1,1)
sns.stripplot(x='cyl',y='hp',data=df)

plt.subplot(2,1,2)
sns.stripplot(x='cyl',y='hp',data=df, jitter=True, size=3)

plt.show()
```

### Constructing swarm plots

```python
plt.subplot(2,1,1)
sns.swarmplot(x='cyl',y='hp', data=df)

plt.subplot(2,1,2)
sns.swarmplot(x='hp',y='cyl',orient='h',hue='origin', data=df)

plt.show()
```

### Constructing violin plots

```python
plt.subplot(2,1,1)
sns.violinplot(x='cyl', y='hp', data=df)

plt.subplot(2,1,2)
sns.violinplot(x='cyl', y='hp', data=df, inner=None, color='lightgray')
sns.stripplot(x='cyl', y='hp', data=df, jitter=True, size=1.5)

plt.show()
```

### Plotting joint distributions

```python
sns.jointplot(x='hp', y='mpg', data=df)
plt.show()
```

```python
sns.jointplot(x='hp', y='mpg', data=df, kind='hex')
plt.show()
```

```python
sns.jointplot(x='hp', y='mpg', data=df, kind='kde')
plt.show()
```

### Plotting distributions pairwise 

```python
auto = df[['mpg','hp','origin']]
auto.head()
```

```python
sns.pairplot(auto)
plt.show()
```

```python
sns.pairplot(data=auto, kind='reg', hue='origin')
plt.savefig('pair.png')
plt.show()
```

### Visualizing correlations with a heatmap

```python
auto_cor = df.corr()
auto_cor.head()
```

```python
sns.heatmap(auto_cor)
plt.show()
```
