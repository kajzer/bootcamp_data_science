### 2

```python
sql_result = pd.read_sql_query("SELECT DISTINCT type, manufacturer, model, seats, engine FROM planes", con)
sql_result.head()
```

```python
planes[['type','manufacturer','model', 'seats', 'engine']].drop_duplicates(['type','manufacturer','model', 'seats', 'engine']).reset_index(drop=True).head()
```

### 3

```python
sql_result = pd.read_sql_query("SELECT COUNT(*), engine FROM planes GROUP BY engine", con)
sql_result.head()
```

```python
pd.DataFrame(planes.groupby('engine', as_index=False).count())[['model','engine']].rename(columns={'model':'COUNT(*)'})
```

```python
pd.DataFrame(planes.groupby('engine', as_index=False).size()).reset_index(drop=False).rename(columns={0:'COUNT(*)'})[['COUNT(*)', 'engine']]
```

```python
pd.DataFrame(pd.value_counts(planes.engine)).reset_index().rename(columns={'index':'engine', 'engine':'COUNT(*)'})[['COUNT(*)','engine']]
```

```python
pd.pivot_table(planes[['engine']],
              index=['engine'],
              aggfunc='size').reset_index()\
                .rename(columns={'engine':'engine', 0:'COUNT(*)'})\
                [['COUNT(*)','engine']]
```

### 4

```python
sql_result = pd.read_sql_query("SELECT COUNT(*), engine, type FROM planes GROUP BY engine", con)
sql_result.head()
```

```python
pd.DataFrame(planes.groupby(['engine', 'type'], as_index=False).size()).reset_index(drop=False).rename(columns={0:'COUNT(*)'})[['COUNT(*)', 'engine', 'type']]
```

```python
pd.pivot_table(planes[['engine', 'type']],
               index=['engine','type'],
              aggfunc='size').reset_index().rename(columns={'engine':'engine', 0:'COUNT(*)'})[['COUNT(*)', 'engine', 'type']]
```

```python
(pd.crosstab(planes.engine, planes.type).unstack().reset_index())
```

```python
planes.groupby(['engine', 'manufacturer'], as_index=False)['year'].agg(['min', 'mean', 'max']).head()
```

```python
pd.pivot_table(planes[['engine', 'year', 'manufacturer']],
              index=['engine', 'manufacturer'],
              aggfunc=['min','max','mean']).head()
```
