### Many plots

```python
theta = np.linspace(0, np.pi*2, 100) # circle
y = 3+2*np.sin(theta)
x = 3+2*np.cos(theta)
plt.plot(x, y, 'k')

x1 = np.array([0, 0, 1, 1])
y1 = np.array([0, 1, 1, 0])
plt.fill(x1, y1, color="#606060")    # square

x2 = np.linspace(-2*np.pi, 0, 50)
y2 = np.sin(x2)+7
plt.plot(x2, y2, color="#404040")    # sin

x3 = np.linspace(-8, -2, 10)
y3 = np.random.normal(3, 1, 10)
plt.scatter(x3, y3, color="#202020") # scatter plot

plt.text(-6, 0, "Aaaaaaargh!")       # text

plt.show()
```

### Legend

```python
x = np.linspace(0, np.pi, 25)
plt.plot(x, np.sin(x**2+np.pi/2), "k-",   label=r"$\sin(x^2_5+\frac{\pi}{2})$")
plt.plot(x, np.cos(x), "rv-.", label="cos(x)")
plt.legend()
plt.show()
```

```python
a1, = plt.plot(x, np.sin(x), "k-")
a2, = plt.plot(x, np.cos(x), "rv-.")
plt.legend([a1, a2], ["sin(x)", "cos(x)"])
plt.show()
```

### Multiple plots

```python
plt.subplot(2, 1, 1)

a1, = plt.plot(x, np.sin(x), "k-")
a2, = plt.plot(x, np.cos(x), "rv-.")
plt.legend([a1, a2], ["sin(x)", "cos(x)"])

plt.subplot(2, 1, 2)

u = np.r_[-5:5:10j]
plt.plot(u, u**2, "rd-.")

plt.show()
```

```python
plt.subplot(1, 2, 1)

a1, = plt.plot(x, np.sin(x), "k-")
a2, = plt.plot(x, np.cos(x), "rv-.")
plt.legend([a1, a2], ["sin(x)", "cos(x)"])

plt.subplot(1, 2, 2)

u = np.r_[-5:5:10j]
plt.plot(u, u**2, "rd-.")

plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

n, bins, patches = plt.hist(x, 20, density=True, facecolor='green', alpha=0.75, edgecolor="k")
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
iris = sns.load_dataset("iris")
```

```python
sns.lmplot(x="sepal_length", y="sepal_width", data=iris, hue='species', fit_reg=False)
plt.axis("equal")
plt.show()
```

```python
sns.lmplot(x="sepal_length", y="sepal_width", data=iris, hue='species', fit_reg=True)
plt.axis("equal")
plt.show()
```

```python
sns.pairplot(iris, diag_kind="kde",
    vars=["sepal_length", "sepal_width", "petal_length", "petal_width"], hue="species")
plt.show()
```

```python
flights=sns.load_dataset("flights")
flights.pivot("month", "year", "passengers")
```

```python
flights_pivot = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_pivot, annot=True, fmt="d", linewidths=0.5)#, annot=True, fmt="d", linewidths=0.5)
plt.show()
```

```python
flights_pivot = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_pivot, annot=False)#, annot=True, fmt="d", linewidths=0.5)
plt.show()
```
