```python
def plot_point(point, angle, length):
     '''
     point - Tuple (x, y)
     angle - Angle you want your end point at in degrees.
     length - Length of the line you want to plot.

     Will plot the line on a 10 x 10 plot.
     '''

     # unpack the first point
     x, y = point

     # find the end point
     endy = y + length * math.sin(math.radians(angle))
     endx = x + length * math.cos(math.radians(angle))

     # plot the points
     ax.set_ylim([0, 10])   # set the bounds to be 10, 10
     ax.set_xlim([0, 10])
     ax.plot([x, endx], [y, endy],'k-')
```

```python
circle1 = plt.Circle((5, 5), 4, facecolor='g', edgecolor='r')
fig, ax = plt.subplots()
ax.add_artist(circle1)
ax.set_xlim((0, 10))
ax.set_ylim((0, 10))
plt.plot([5, 5], [2, 1], 'k-')
plt.plot([8, 9], [5, 5], 'k-')
plt.plot([1, 2], [5, 5], 'k-')
plt.plot([5, 5], [9, 8], 'k-')
plt.text(5,7,'12',ha='center')
plt.text(5,3,'6',ha='center',va='center')
plt.text(7.5,5,'3',ha='center',va='center')
plt.text(2.5,5,'9',ha='center',va='center')
degree = 0.5 * (60*int(now.strftime('%I'))+now.minute)
plot_point((5,5), degree+180, 2)
degree2 = 6*now.minute
plot_point((5,5), degree2+45, 3)
plt.show()
```
