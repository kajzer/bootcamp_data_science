```python
df2 = df.copy()
pattern = r"Z(\d+)\((\d+)\+(\d+)\)"
df2.rename(columns={"Z13(5)": "Z13(4+1)"}, inplace=True)
```

```python
for col in df2.columns:
    if '+' in col:
        name_extr = re.findall(pattern, col)
        col0_name = 'Z'+name_extr[0][0]+'(' + name_extr[0][1] + ')'
        col1_name = 'D'+name_extr[0][0]+'(' + name_extr[0][2] + ')'
        df2['str_split'] = df2[col].str.split('+')
        df2[col0_name] = df2.str_split.str.get(0)
        df2[col1_name] = df2.str_split.str.get(1)
        del df2[col]
del df2['str_split']
del df2['Z15(5)']
```

```python
df2
```

```python
def recode_na_nb_ns(value):
    if 'NA' in value or 'NS' in value or 'NB' in value:
        return '0'
    if 'ZW' in value or 'US' in value:
        return np.NaN
    else:
        return value

for col in df2.columns:
    if 'Z' in col or 'D' in col:
        df2[col] = df2[col].apply(recode_na_nb_ns)

for col in df2.columns:
    if 'Z' in col or 'D' in col:
        df2[col]=df2[col].astype(float)
```
