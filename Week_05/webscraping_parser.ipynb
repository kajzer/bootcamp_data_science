# Webscraping with parser

```python
from bs4 import BeautifulSoup
import io, os.path, re
import requests, urllib, json
import numpy as np
import pandas as pd
```

```python
response = requests.get('https://stooq.pl/q/?s=cdr')
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
links = soup.findAll('a')
for link in links:
    print(link.text)
```

```python
for link in links:
    print(link['href'])
```

```python
assert len(soup.find_all('a')) < len(soup.select("[href]"))
```

```python
links = soup.select("[href]")
```

```python
for link in links:
    print(link)
```

```python
links = soup.find_all('a', href=True)
for link in links:
    print(link.text)
    print(link.get('href'))
    print('----------')
```

```python
response = requests.get("https://www.filmweb.pl/search?q=avengers")
soup = BeautifulSoup(response.text, 'html.parser')
soup.find_all('a', class_='filmPreview__link')[0]['href']
```

```python
title = "avengers"
f = { 'q' : title }
response = requests.get('https://www.filmweb.pl/search', params=f)
soup = BeautifulSoup(response.text, 'html.parser')
for tag in soup.find_all(class_="filmPreview__title"):
    title_tag = tag.text.lower()
    if title.lower() == title_tag:
        print(title_tag)
        print(tag.parent['href'])
```
